***What is a t-test?***

A t-test is a statistical method used as a means to compare between groups or to see if a sample piece of data is different than a known value. The t-test is used when the standard deviation is not known and the data comes from a normally distributed dataset. 
Types of t-tests:
1.	One-sample t-test: Compares the mean of a single sample with a known value.
2.	Paired-t-test: Evaluate the mean difference two related groups.
3.	Independent t-test- Compares two independent groups. 



When a t-test is performed in python, it returns two result; P-Value and statistic. The P-value greater than 0.05 indicates a significant difference.

***Perform a t-test to determine whether there is a significant difference between the two treatment groups trt1 and trt2.***

```python
#Ctrl
ctrl = np.array([4.17, 5.58, 5.18, 6.11, 4.5, 4.61, 5.17, 4.53, 5.33, 5.14])

#Trt1
trt1 = np.array([4.81, 4.17, 4.41, 3.59, 5.87, 3.83, 6.03, 4.89, 4.32, 4.69])

#Trt2 
trt2 = np.array([6.31, 5.12, 5.54, 5.5, 5.37, 5.29, 4.92, 6.15, 5.8, 5.26])


test = stats.ttest_rel(trt1, trt2)
print (test)
```

**Conclusion**:

Given the t-statistic as -2.84, and the p-value as 0.0192031, I believe you reject the null hyprothesis. This means theres a difference between the two groups. The negative t-statistic that trt1 has a lower mean than trt2.

***What is an ANOVA?***

ANOVA, or Analysis of Variance, is a statistical method used to '...determine differences between research results from three or more unrelated samples or groups." (https://www.qualtrics.com/en-gb/experience-management/research/anova/)

***Performing ANOVA on PlantData***

```python
result = _, p_value = stats.f_oneway(ctrl, trt1, trt2)

print (result)
```

**Conclusion**

Once again we can reject the null hypothesis as this the number is lower than 0.05. 

***Why it is more appropriate to apply ANOVA rather than several t-tests when analyzing more than two groups?***

Within the realm of statitsical analysis, when comparing two or more groups in an experiment or study. It is important to chose the right test to get reliable results. As mentioned earlier, when comparing two groups, one uses a t-test and when comparing two or more groups, an Anova tests is best. The Anova, or analysis of variance, is more appropriate to analysing more than two groups, as opposed to several t-tests, for the follwoing reasons: 
  

1. Using multiple t-tests leads to increased type 1 errors. Type 1 errors happen when the test incorrectly rejects a null hypothesis and therefore wrongly says that there is a difference between the grops even though there is not. According to the blog from Voxco, several t-tests increases the chance of a false positive with each t-tes having a 5% probability of prouding a Type 1 error. This increases with the numebr of tests perfomred. Anova minimizes the risk of finding these errors by considering all data in one calculation.

2. Anova is more efficient. As a whole, Anova is more effient and can evaluate differences among several groups in a single undertaking, saving time. Instead of dealing in paits of groups, Anova, deals with groups, giving faster and cleaer answers. Anova is designed specifically to deal with three or more groups making it better than the several t-tests approach.


In conclusion, when comparing more than two groups, ANOVA is the preferred method over multiple t-tests. ANOVA controls for Type I error rates, is more efficient, and can handle comparisons of multiple groups.
