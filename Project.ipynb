***Describe what a t-test is, how it works, and what the assumptions are.***

A t-test is a statistical method used as a means to compare between groups or to see if a sample piece if data is different than a known value. The t-test is used when the standard deviation is not known and the data comes from a normally distributed dataset. 
Types of t-tests:
1.	One-sample t-test: Compares the mean of a single sample with a known value.
2.	Paired-t-test: Evaluate the mean difference two related groups.
3.	Independent t-test- Compares two independent groups. 

Below is an example of a Paired t-test I completed in my tasks.ipynb: 


```python
import numpy as np
import scipy.stats as stats

#Data 
before = np.array([63,68,70,64,74,67,70,57,66,65])

after = np.array([64,64,68,64,73,70,72,54,61,63])

#t-test
test = stats.ttest_rel(before, after)
print (test) 
```

This begins with the libraires needed. Scipy.stats is a statistical function where ttest_rel is used to do a paired t-test. 
The #data section then creates two arrays, before and after. 
Next the #t-test section performs the t-test and compares the differences between each pari to determine if the mean difference is different from 0. The results are then printed.

The printed results show:

1.	T-Statistic (1.33): shows the amount of difference with a larger amount suggesting a bigger difference. 
2.	P-Value (0.21)- if the p-value is less than 0.05, the null hypothesis is rejected, and it is assumed there is a significant difference between the datas means. If it is greater than 0.05 the data is inconclusive and there isnt enough evidence to show a difference. This result is less than 0.05 so the you can reject the null hypothesis.

***Perform a t-test to determine whether there is a significant difference between the two treatment groups trt1 and trt2.***

***Perform ANOVA to determine whether there is a significant difference between the three treatment groups ctrl, trt1, and trt2.***

```python
import numpy 
import scipy.stats as stats

#Ctrl
ctrl = np.array([4.17, 5.58, 5.18, 6.11, 4.5, 4.61, 5.17, 4.53, 5.33, 5.14])

#Trt1
trt1 = np.array([4.81, 4.17, 4.41, 3.59, 5.87, 3.83, 6.03, 4.89, 4.32, 4.69])

#Trt2 
trt2 = np.array([6.31, 5.12, 5.54, 5.5, 5.37, 5.29, 4.92, 6.15, 5.8, 5.26])


_, p_value = stats.f_oneway(ctrl, trt1, trt2)

print(f"P-Value: {p_value}")
```
